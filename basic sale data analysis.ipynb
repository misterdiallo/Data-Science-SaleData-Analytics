#### Visualization 


```python
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# set the width of our figure
plt.figure(figsize=(16,6))

# set title for our figure
plt.title("Sales Visualization  ")

# giving data to visualize
sns.lineplot(data=clean_sales_2)
```

```python
# Subset of data
list(clean_sales_2)
```

```python
# Linear diagram with all the product lines 
sns.lineplot(data=clean_sales_2["PRODUCTLINE"], label="PRODUCT LINE") 
# adding label for horizontal axis
plt.xlabel("Date")
```

```python
# Bar chart with all the product lines with sales volume
plt.figure(figsize=(10,6))
plt.title("Bar Chart visualization")
sns.barplot(x=clean_sales_2.PRODUCTLINE, y=clean_sales_2["SALES"]) 
plt.ylabel("Sales volume")
```

```python
# 
sns.scatterplot(x=clean_sales_2['PRODUCTLINE'], y=clean_sales_2["STATE"], hue=clean_sales_2["TERRITORY"])
```

```python
# Histogramm
sns.histplot(data=clean_sales_2, x="PRICEEACH", hue="TERRITORY")
plt.title("Histogramm")
```

```python
sns.histplot(clean_sales_2['MSRP'])
```

```python
sns.kdeplot(data=clean_sales_2, x="PRICEEACH", hue="TERRITORY", shade=True, warn_singular=False)
plt.title("Area Graph ")
```
